# 3. DATA PRE-PROCESSING

## 3.1 MANAGE NULL VALUES

```python
df = df.drop(['Wake_Up_Time',
              'Bed_Time',
              'Occupation',
              'Gender',
              'Marital_Status',
              'Exercise_Type',
              'Meditation_Practice'], axis=1)
```

```python
df['Blood_Pressure'].mean(), df['Blood_Pressure'].median(), df['Blood_Pressure'].mode()
```

```python
df.info()
```

```python
df['Screen_Time'] = df['Screen_Time'].astype('float64')
df['Travel_Time'] = df['Travel_Time'].astype('float64')
df['Physical_Activity'] = df['Physical_Activity'].astype('float64')
df['Sleep_Duration'] = df['Sleep_Duration'].astype('float64')
```

```python
df.info()
```

```python
def convertObjToInt(cols):
    for i in cols:
        data = pd.DataFrame({i: df[i].unique()})
        data_LE = LabelEncoder()
        data_LE.fit(np.ravel(data))
        df[i] = data_LE.transform(df[i])

columns = ['Stress_Detection', 'Smoking_Habit']
convertObjToInt(columns)
```

# 4. TRAIN - TEST SPLITTING 

```python
import pandas as pd
df = pd.read_csv('cleaned_data.csv')

df.columns
```

```python
X = df.drop('Stress_Detection', axis=1)   
y = df['Stress_Detection']                
```

```python
from sklearn.model_selection import train_test_split
train_X, test_X, train_y, test_y = train_test_split(X,y, test_size=0.3, shuffle=False) # 80:20, 70:30, 60:40
```

```python
print(train_X.shape, train_y.shape)
print(test_X.shape, test_y.shape)
```

# 5. APPLYING CLASSIFIER

## 5.1  Support Vector Classifier

```python
from sklearn.svm import SVC # Support Vector Classifier

model_svc = SVC()
model_svc.fit(train_X, train_y)

print(model_svc)
```

```python
model_pred_svc = model_svc.predict(test_X)
model_pred_svc
```

```python
from sklearn.metrics import accuracy_score
model_acc_svc = accuracy_score(test_y, model_pred_svc)
print(round(model_acc_svc,3)*100, "%")
```
